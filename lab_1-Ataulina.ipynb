```python
import random

n = 20

def generate_matrix():  
    return [[random.randrange(-1000, 1000) for y in range(n)] for x in range(n)]

matrix = generate_matrix()
for i in range(n):
    print(matrix[i])
```

### Сортировка встроенная

```python
def DS(matrix):
    for i in range(n):
        default_sort(matrix[i])
    return matrix
def default_sort(arr):
    arr.sort()
    return arr

matrix2 = DS(matrix)
for i in range(n):
    print(matrix2[i])
```

### Сортировка выбором

```python
def SS(matrix):
    for i in range(n):
        selection_sort(matrix[i])
    return matrix
def selection_sort(arr):
    for i in range(0, len(arr) - 1):
        smallest = i
        for j in range(i + 1, len(arr)-1):
            if arr[j] < arr[smallest]:
                smallest = j
        arr[i], arr[smallest] = arr[smallest], arr[i]
    return arr
matrix2 = SS(matrix)
for i in range(n):
    print(matrix2[i])
```

### Сортировка вставкой

```python
def IS(matrix):
    for i in range(n):
        insertion_sort(matrix[i])
    return matrix
def insertion_sort(arr):
    for i in range(1, len(arr)):
        temp = arr[i]
        j = i - 1
        while (j >= 0 and temp < arr[j]):
            arr[j + 1] = arr[j]
            j = j - 1
        arr[j + 1] = temp
    return arr
matrix2 = IS(matrix)
for i in range(n):
    print(matrix2[i])
```

### Сортировка обменом

```python
def BS(matrix):
    for i in range(n):
        bubble_sort(matrix[i])
    return matrix
def bubble_sort(arr):
    for main_index in range(len(arr) - 1):
        for index in range(0, len(arr) - main_index - 1):
            if arr[index] > arr[index + 1]:
                arr[index], arr[index + 1] = arr[index + 1], arr[index]
    return arr
matrix2 = BS(matrix)
for i in range(n):
    print(matrix2[i])
```

### Сортировка Шелла

```python
def ShS(matrix):
    for i in range(n):
        shell_sort(matrix[i])
    return matrix
def shell_sort(arr):
    step = len(arr) // 2 # ищем шаг
    while step > 0:
        for i in range(step, len(arr)):
            value = arr[i]
            j = i
            while j >= step and arr[j - step] > value:
                arr[j] = arr[j - step]
                j = j - step
            arr[j] = value
        step = step // 2
    return arr
matrix2 = ShS(matrix)
for i in range(n):
    print(matrix2[i])
```

### Быстрая сортировка

```python
def QS(matrix):
    for i in range(n):
        quicksort(matrix[i], 0, len(matrix))
    return matrix
def quicksort(arr, start, end):
    if end - start > 1:
        p = partition(arr, start, end)
        quicksort(arr, start, p)
        quicksort(arr, p + 1, end)

def partition(arr, start, end):
    pivot = arr[start]
    i = start + 1
    j = end - 1
 
    while True:
        while (i <= j and arr[i] <= pivot):
            i = i + 1
        while (i <= j and arr[j] >= pivot):
            j = j - 1
 
        if i <= j:
            arr[i], arr[j] = arr[j], arr[i]
        else:
            arr[start], arr[j] = arr[j], arr[start]
            return j
matrix2 = QS(matrix)
for i in range(n):
    print(matrix2[i])
```

### Турнирная сортировка

```python
def TS(matrix):
    for i in range(n):
        tournament_sort(matrix[i])
    return matrix
def tournament_sort(arr):
    for i in range(len(arr)):
        a = 1
        while a < len(arr):
            k = 0
            while k*a*2 < len(arr):
                if k*a*2+a < len(arr):
                    if arr[k*a*2 + a] < arr[k*a*2]:
                        arr[k*a*2 + a], arr[k*a*2] = arr[k*a*2], arr[k*a*2+a]
                k+=1
            a*=2
    return arr
matrix2 = TS(matrix)
for i in range(n):
    print(matrix2[i])
```

### Пирамидальная сортировка

```python
def HS(matrix):
    for i in range(n):
        heap(matrix[i])
    return matrix
def heapify(heap_array, n, k):  # Heap sort helper function
    largest = k
    left = 2 * k + 1
    right = 2 * k + 2
    if left < n and heap_array[largest] < heap_array[left]:
        largest = left
    if right < n and heap_array[largest] < heap_array[right]:
        largest = right
    if largest != k:
        heap_array[k], heap_array[largest] = heap_array[largest], heap_array[k]
        heapify(heap_array, n, largest)
    return heap_array

def heap(str_array):  # Heap sort function
    for i in range(len(str_array) - 1, -1, -1):
        heapify(str_array, len(str_array), i)
    for i in range(len(str_array) - 1, 0, -1):
        str_array[i], str_array[0] = str_array[0], str_array[i]
        heapify(str_array, i, 0)
    return str_array
matrix2 = HS(matrix)
for i in range(n):
    print(matrix2[i])
```
